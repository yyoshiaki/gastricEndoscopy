```python
import os
from PIL import Image, ImageOps, ImageFilter
import numpy as np
import matplotlib.pyplot as plt
import skimage
%matplotlib inline
```

```python
def process_img_AG(dir_img_AG, out_dir_img_AG):
    files_AG = os.listdir(dir_img_AG)
    files_AG = [x for x in files_AG if x[0] != '.']
    for i in range(len(files_AG)):
        img = Image.open(dir_img_AG+'/'+files_AG[i])
        w, h = img.size
        s = min(w,h)

        if w > h:
            t = 0
            b = s

            l = (w - s) / 2
            r = l + s

            box = (l, t, r, b)
        else:
            t = 0
            r = s

            l = (h - s) / 2
            b = t + s

            box = (l, t, r, b)

        img = img.crop(box)
        thumbnail_size = (150, 150)
        img.thumbnail(thumbnail_size)
        # print(files_AG[i])
        img.convert("RGB").save(out_dir_img_AG + '/{}.jpg'.format(i), 'JPEG')

    for i in range(len(files_AG)):
        img = Image.open(dir_img_AG+'/'+files_AG[i])
        w, h = img.size
        s = min(w,h)

        if w > h:
            t = 0
            b = s

            l = (w - s) / 2
            r = l + s

            box = (l, t, r, b)
        else:
            t = 0
            r = s

            l = (h - s) / 2
            b = t + s

            box = (l, t, r, b)

        img = img.crop(box)
        thumbnail_size = (150, 150)
        img.thumbnail(thumbnail_size)
        # print(files_AG[i])

        img = img.rotate(90)
        img = img.point(lambda x: x*1.3)

        i += len(files_AG)
        img.convert("RGB").save(out_dir_img_AG + '/{}.jpg'.format(i), 'JPEG')

    for i in range(len(files_AG)):
        img = Image.open(dir_img_AG+'/'+files_AG[i])
        w, h = img.size
        s = min(w,h)

        if w > h:
            t = 0
            b = s

            l = (w - s) / 2
            r = l + s

            box = (l, t, r, b)
        else:
            t = 0
            r = s

            l = (h - s) / 2
            b = t + s

            box = (l, t, r, b)

        img = img.crop(box)
        thumbnail_size = (150, 150)
        img.thumbnail(thumbnail_size)
        # print(files_AG[i])

        img = ImageOps.flip(img)
        img = img.point(lambda x: x*(1/1.3))

        i += 2*len(files_AG)
        img.convert("RGB").save(out_dir_img_AG + '/{}.jpg'.format(i), 'JPEG')
    
    for i in range(len(files_AG)):
        img = Image.open(dir_img_AG+'/'+files_AG[i])
        w, h = img.size
        s = min(w,h)

        if w > h:
            t = 0
            b = s

            l = (w - s) / 2
            r = l + s

            box = (l, t, r, b)
        else:
            t = 0
            r = s

            l = (h - s) / 2
            b = t + s

            box = (l, t, r, b)

        img = img.crop(box)
        thumbnail_size = (150, 150)
        img.thumbnail(thumbnail_size)
        # print(files_AG[i])

        img = img.rotate(270)
        imgarr = skimage.util.random_noise(np.array(img), mode='gaussian', seed=0, var=0.002)
        img = Image.fromarray(np.uint8(imgarr*255))

        i += 3*len(files_AG)
        img.convert("RGB").save(out_dir_img_AG + '/{}.jpg'.format(i), 'JPEG')
```

```python
def process_img_norm(dir_img_norm, out_dir_img_norm):
    files_norm = os.listdir(dir_img_norm)
    files_norm = [x for x in files_norm if x[0] != '.']
    for i in range(len(files_norm)):
        img = Image.open(dir_img_norm+'/'+files_norm[i])
        w, h = img.size
        s = min(w,h)

        if w > h:
            t = 0
            b = s

            l = (w - s) / 2
            r = l + s

            box = (l, t, r, b)
        else:
            t = 0
            r = s

            l = (h - s) / 2
            b = t + s

            box = (l, t, r, b)

        img = img.crop(box)
        thumbnail_size = (150, 150)
        img.thumbnail(thumbnail_size)
        img.convert("RGB").save(out_dir_img_norm + '/{}.jpg'.format(i), 'JPEG')

    for i in range(len(files_norm)):
        img = Image.open(dir_img_norm+'/'+files_norm[i])
        w, h = img.size
        s = min(w,h)

        if w > h:
            t = 0
            b = s

            l = (w - s) / 2
            r = l + s

            box = (l, t, r, b)
        else:
            t = 0
            r = s

            l = (h - s) / 2
            b = t + s

            box = (l, t, r, b)

        img = img.crop(box)
        thumbnail_size = (150, 150)
        img.thumbnail(thumbnail_size)

        img = img.rotate(90)
        img = img.point(lambda x: x*1.3)

        i += len(files_norm)

        img.convert("RGB").save(out_dir_img_norm + '/{}.jpg'.format(i), 'JPEG')

    for i in range(len(files_norm)):
        img = Image.open(dir_img_norm+'/'+files_norm[i])
        w, h = img.size
        s = min(w,h)

        if w > h:
            t = 0
            b = s

            l = (w - s) / 2
            r = l + s

            box = (l, t, r, b)
        else:
            t = 0
            r = s

            l = (h - s) / 2
            b = t + s

            box = (l, t, r, b)

        img = img.crop(box)
        thumbnail_size = (150, 150)
        img.thumbnail(thumbnail_size)

        img = ImageOps.flip(img)
        img = img.point(lambda x: x*(1/1.3))

        i += 2*len(files_norm)
        img.convert("RGB").save(out_dir_img_norm + '/{}.jpg'.format(i), 'JPEG')
        
    for i in range(len(files_norm)):
        img = Image.open(dir_img_norm+'/'+files_norm[i])
        w, h = img.size
        s = min(w,h)

        if w > h:
            t = 0
            b = s

            l = (w - s) / 2
            r = l + s

            box = (l, t, r, b)
        else:
            t = 0
            r = s

            l = (h - s) / 2
            b = t + s

            box = (l, t, r, b)

        img = img.crop(box)
        thumbnail_size = (150, 150)
        img.thumbnail(thumbnail_size)

        img = img.rotate(270)
        imgarr = skimage.util.random_noise(np.array(img), mode='gaussian', seed=0, var=0.002)
        img = Image.fromarray(np.uint8(imgarr*255))

        i += 3*len(files_norm)
        img.convert("RGB").save(out_dir_img_norm + '/{}.jpg'.format(i), 'JPEG')
```

```python
process_img_AG('dataset/raw_images/img_AG_new/', 'dataset/train_images/AtrophicGastritis/')
process_img_AG('dataset/raw_images/img_AG_new_test/', 'dataset/test_images/AtrophicGastritis/')
process_img_norm('dataset/raw_images/img_norm_new/', 'dataset/train_images/normal/')
process_img_norm('dataset/raw_images/img_norm_new_test/', 'dataset/test_images/normal/')
```

```python
def deploy_img(dir_img, f_out):
    files = ['{}.jpg'.format(i) for i in range(10)]
    plt.figure(dpi=150, figsize=(20,8))
    for i,f in enumerate(files):
        img = Image.open(dir_img+f)
        plt.subplot(2,5,i+1)
        plt.title(f)
        plt.imshow(np.array(img))
    plt.savefig(f_out)
```

```python
deploy_img('dataset/train_images/AtrophicGastritis/',  'AG_imgs.pdf')
```
